```python
data = data.drop("Last Update", axis = 1)
```

```python
data.describe()
```

```python
data.head()
```

```python
# Filter data for years 2020 to 2024
data = data[(data['Year'] >= 2020) & (data['Year'] <= 2024)]

# Check the filtered data
print(data.head())
```

```python
# Get unique states
unique_states = data['State'].unique()
print(unique_states)
```

```python
# Total industry count per state (from 2020 to 2024)
state_wise_count = data.groupby('State')['Count'].sum().reset_index()
print(state_wise_count)
```

```python
# Top 5 industries with the highest presence
top_industries = data.groupby('Industry')['Count'].sum().nlargest(5).reset_index()
print(top_industries)
```

```python
data.info()
```

```python
# State with the most diverse industries
state_diversity = data.groupby('State')['Industry'].nunique().reset_index().sort_values(by='Industry', ascending=False)
print(state_diversity.head(1))  # Top 1 most diverse state
```

```python
# Year-wise total industry growth
yearly_growth = data.groupby('Year')['Count'].sum().reset_index()
print(yearly_growth)
```

```python
# Most popular industry in each state
most_popular_industry = data.groupby(['State', 'Industry'])['Count'].sum().reset_index()
most_popular_industry = most_popular_industry.loc[most_popular_industry.groupby('State')['Count'].idxmax()]
print(most_popular_industry)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure your data is sorted by Year
data = data.sort_values(by='Year')

# Unique industries
industries = data['Industry'].unique()

# Generate line plots for each industry
for industry in industries:
    industry_data = data[data['Industry'] == industry]
    plt.figure(figsize=(8, 4))
    sns.lineplot(x='Year', y='Count', data=industry_data, marker='o')
    plt.title(f'Trend for {industry}')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()
```
